```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Data
monthly_expenses = [
    {'Category': 'Housing', 'Amount': random.randint(1000, 3000)},
    {'Category': 'Utilities', 'Amount': random.randint(100, 500)},
    {'Category': 'Groceries', 'Amount': random.randint(200, 800)},
    {'Category': 'Entertainment', 'Amount': random.randint(50, 300)},
    {'Category': 'Dining Out', 'Amount': random.randint(100, 400)}
    # Add more categories as needed
]

# Extract data
categories = [expense['Category'] for expense in monthly_expenses]
amounts = [expense['Amount'] for expense in monthly_expenses]

# Create a stacked histogram for monthly expenses
plt.bar(categories, amounts, color='skyblue')

# Customize the plot with labels, title, and legend
plt.xlabel('Expense Categories')
plt.ylabel('Amount ($)')
plt.title('Monthly Expenses by Category')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Data
exam_scores = {
    'Subjects': ['Math', 'Science', 'History'],
    'Group_A': [random.randint(60, 100) for _ in range(3)],
    'Group_B': [random.randint(60, 100) for _ in range(3)]
}

# Extract data
subjects = exam_scores['Subjects']
scores_group_a = exam_scores['Group_A']
scores_group_b = exam_scores['Group_B']

# Create a grouped bar chart for exam scores comparison
bar_width = 0.35  # Width of the bars
bar_positions_group_a = np.arange(len(subjects))
bar_positions_group_b = bar_positions_group_a + bar_width

plt.bar(bar_positions_group_a, scores_group_a, width=bar_width, label='Group A', alpha=0.7)
plt.bar(bar_positions_group_b, scores_group_b, width=bar_width, label='Group B', alpha=0.7)

# Customize the plot with labels, title, and legend
plt.xlabel('Subjects')
plt.ylabel('Exam Scores')
plt.title('Exam Scores Comparison - Group A vs. Group B')
plt.xticks(bar_positions_group_a + bar_width / 2, subjects)
plt.legend()

# Show the plot
plt.show()
```
